### Importing relevant packages

```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
```

### Frequentist approach

#### Generating random Bernoulli experiments with different sample sizes

```python
# creating theta values from 0 to 1
theta=np.linspace(0,1,100)
##########################################
n=10
m=9
likelihood1= (theta**m) * (1-theta)**(n-m)
likelihood1=likelihood1/sum(likelihood1)
##########################################
n=100
m=90
likelihood2= (theta**m) * (1-theta)**(n-m)
likelihood2=likelihood2/sum(likelihood2) 
#########################################
n=1000
m=900
likelihood3= (theta**m) * (1-theta)**(n-m)
likelihood3=likelihood3/sum(likelihood3) 
```

### Visualize confidence in results with increasing sample size

```python
plt.figure(figsize=(10,6))
plt.plot(theta,likelihood1,label='sample_size=10')
plt.legend()
plt.plot(theta,likelihood2,label='sample_size=100')
plt.legend()
plt.plot(theta,likelihood3,label='sample_size=1000')
plt.legend()
plt.title('maximum likelihood estimation')
plt.grid()
plt.show()
```

#### The above graph shows that probability mass concentrates with larger samples which in turns provides higher confidence in results

### Bayesian approach

```python
n=10
m=9

# creating theta vector from 0 to 1
theta=np.linspace(0,1,100)

# likelihood
likelihood= (theta**m) * (1-theta)**(n-m)
likelihood=likelihood/sum(likelihood)

# prior
prior= beta.pdf(theta,3,7)
prior= prior/sum(prior)

# posterior
posterior= likelihood*prior
posterior=posterior/sum(posterior)
```

```python
plt.figure(figsize=(10,6))
plt.plot(theta,prior,label='prior',c='r')
plt.legend()
plt.plot(theta,likelihood,label='likelihood',c='b')
plt.legend()
plt.plot(theta,posterior,label='posterior',c='black')
plt.legend()

plt.title('maximum a posteriori')
plt.grid()
plt.show()
```

#### it's very clear now how prior drags posterior towards lower theta values
